```python
from langchain.tools import tool
```

```python
@tool
def multiply(a: int, b: int) -> int:
    """ multiply two integers
    args:
        a: first integer
        b: second integer
    returns:
        integer: the result of multiplying a and b"""
    return a * b
```

```python
def getWeather(city: str) -> str:
    """Get the current weather in a city."""
    return f"The current weather in {city} is sunny."
```

```python
weather_tool = StructuredTool.from_function(
    func=getWeather,
    name = "getWeather",
    description = "Get the current weather in a city.",
    args_schema = WeatherInput # now this model will be used
)
```

```python
class WeatherInput(BaseModel):
    city: str = Field(..., description="The city to get the weather for")
    units: str = Field("metric", description="The units for temperature (metric or imperial)")

class GetWeatherTool(StructuredTool):
    name = "getWeather"
    description = "Get the current weather in a city."
    args_schema = WeatherInput

    def _run(self, city: str, units: str = "metric") -> str:
        """Get the current weather in a city."""
        return f"The current weather in {city} is sunny with a temperature of 25 degrees {units}."  
```
